```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('/kaggle/input/opencv-samples-images/data/messi5.jpg')
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)
```

```python
image_ = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("gray")
plt.imshow(image_,cmap='gray')
```

```python
image = cv2.line(image,(0,0),(350,500),(250,0,0),9)
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("w/ line")
plt.imshow(image)
```

```python
# Negative image
image2 = abs(255-image)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,20))

plt.subplot(1,2,1)
plt.title("negative")
plt.imshow(image2)
```

```python
image3 = cv2.imread('/kaggle/input/opencv-samples-images/minions.jpg') 
```

```python
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image3)
```

```python
image3=  cv2.circle(image3, (450,400), 300, (0,0,255),20)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image3)
```

```python
image4 = cv2.imread('/kaggle/input/opencv-samples-images/data/apple.jpg', 0)
image4 = cv2.cvtColor(image4,  cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image4)
```

```python
image4 = cv2.rectangle(image4, (0,0),(500,500),(255,0,0),20)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image4)
```

## learn to make a black screen

```python
image5 = np.zeros((512,512),np.uint8)
image5 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(image5)
```

## Arithmetic operations

```python
image6 = cv2.rectangle(image5,(100,100),(200,200),(255,255,255),-1)

plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image6)
```

```python
image7 = cv2.rectangle(image5,(200,100),(300,200),(255,255,255),-1)

plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image7)
```

```python
weightedSum = cv2.addWeighted(image6, 0.5, image7, 0.5, 0)
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(weightedSum)
```
